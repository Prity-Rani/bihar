```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
import matplotlib.pyplot as plt
import json
from collections import OrderedDict
```

```python
def trend(consti_number):
    url="https://results.eci.gov.in/ACTRENDS2020/"
    a="ConstituencywiseS04"
    #consti_number=str(consti_number)
    b=str(consti_number)
    c=".htm?ac="
    url=url+a+b+c+b
    response=req.get(url)
    soup=bs(response.text)
    table=soup.find_all("tr",{"style":"font-size:12px;"})
    row=[]
    for i in table:
        z=i.find_all("td")
        row_item={"osn":z[0].text,
                "candidate":z[1].text,
                "party":z[2].text,
                "evm_votes":z[3].text,
                "postal_votes":z[4].text,
                "total_votes":z[5].text,
                "percentage_votes":z[6].text,
                "contituencty":b}
        row.append(row_item)
    df=pd.DataFrame(row)
    return df
```

```python
frames=[]
for i in range(1,244):
    frames.append(trend(i))
    time.sleep(0.25)
result=pd.concat(frames)
result
```

```python
result.to_csv("bihar_2020.csv")
```
